# ðŸ“Š Superstore Sales Analysis

This notebook loads and explores the Sample Superstore dataset.

```python

import pandas as pd

# Load data
df = pd.read_csv('../data/Sample - Superstore.csv', encoding='ISO-8859-1')

# Preview data
df.head()
```

## Quick Data Overview

```python

# Summary
df.info()
df.describe(include='all')
```

## Next Steps
- Group sales by category and region
- Visualise top sales by segment
- Build interactive dashboard in Power BI

```python
# Group sales by Category and Region
grouped = df.groupby(['Category', 'Region']).agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum'
}).reset_index()

# Display grouped data
grouped.sort_values('Sales', ascending=False)
```

```python
import matplotlib.pyplot as plt

# Plot total sales by category
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,6))
category_sales.plot(kind='bar', color='#E86130')
plt.title('Total Sales by Category')
plt.ylabel('Sales')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Convert Order Date to datetime (if not already)
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Create Year-Month column
df['YearMonth'] = df['Order Date'].dt.to_period('M').astype(str)

# Group sales by Year-Month
monthly_sales = df.groupby('YearMonth')['Sales'].sum().reset_index()

# Plot Monthly Sales
plt.figure(figsize=(12,6))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Sales'], marker='o', color='#E86130')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Save grouped data for Power BI
grouped.to_csv('../data/superstore_grouped.csv', index=False)
monthly_sales.to_csv('../data/superstore_monthly_sales.csv', index=False)
```
